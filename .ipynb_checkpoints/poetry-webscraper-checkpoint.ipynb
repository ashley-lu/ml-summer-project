```python
import pandas as pd
import requests
import re
import urllib3
from bs4 import BeautifulSoup as bsoup
```

```python
page = requests.get("https://www.poetryfoundation.org/poems/browse#page=1")
soup = bsoup(page.content, "lxml")
poems = soup.find_all('a',href=re.compile('.*/poems/[0-9]+/.*'))
```

```python
poems = []
```

```python
for i in range(1, 2302):
    page = requests.get("https://www.poetryfoundation.org/poems/browse#page={0}".format(i))
    soup = bsoup(page.content, "lxml")
    poems.append(soup.find_all('a',href=re.compile('.*/poems/[0-9]+/.*')))
```

```python
poems
```

```python
title = []
author = []
text = []
```

```python
poem = poems[3]
poem_link = poem['href']
http = urllib3.PoolManager()
response = http.request('GET', poem_link)
poem_soup = bsoup(response.data.decode('utf-8'))
poem_soup
```

```python
poem_title = poem_soup.find('h1').get_text().strip()
re.sub(r'\n\s*\n', r'\n\n', poem_title, flags=re.M)
poem_title
```

```python
poem_author = poem_soup.find('a', href=re.compile('https://www.poetryfoundation.org/poets/.*')).get_text()
poem_author
```

```python
for poem in poems:
    poem_link = poem['href']
    # poem_page = urllib2.urlopen(poem_link)
    http = urllib3.PoolManager()
    response = http.request('GET', poem_link)
    poem_soup = bsoup(response.data.decode('utf-8'))
    # poem_soup = bsoup(poem_page.read())
    
    poem_title = poem_soup.find('h1')
    if poem_title:
        poem_title = poem_title.text.encode('utf8')
        title.append(poem_title)
    else:
        continue
    poem_author = poem_soup.find('a', href="https://www.poetryfoundation.org/poets/.*")
    author.append(poem_author)
    poem_text = poem_soup.find('p')
    text.append(poem_text)
```
